```python
import warnings
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Suppress specific DeprecationWarnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

columns = ['CountyName', 'DistrictName',
       'SchoolName', 'ClassID', 'CourseCode']

# Step 1: Focus on 'districtwide' and 'countyname' columns
df_filtered = dataset_to_use[columns]

# Step 1: Convert each row to a transaction format
transactions = df_filtered.apply(lambda x: list(x.dropna().astype(str)), axis=1).tolist()

# Step 2: One-hot encode the transactions
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
transaction_df = pd.DataFrame(te_ary, columns=te.columns_)

# Step 3: Apply FP-Growth to find frequent itemsets
frequent_itemsets = fpgrowth(transaction_df, min_support=0.0001, use_colnames=True)

# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.0001)

# Step 5: Create a dictionary to track correlated attribute sets for each column
correlated_attributes = {col: set() for col in columns}

# Step 6: Populate correlated attributes for each column
for _, row in rules.iterrows():
    antecedents = list(row['antecedents'])
    consequents = list(row['consequents'])

    # Check if antecedents and consequents are from the same column
    for col in columns:
        col_antecedents = [item for item in antecedents if item in dataset_to_use[col].astype(str).unique()]
        col_consequents = [item for item in consequents if item in dataset_to_use[col].astype(str).unique()]

        # If the antecedents and consequents are in the same column, consider them correlated
        if col_antecedents and col_consequents:
            correlated_attributes[col].update(col_antecedents + col_consequents)

# Step 7: Output correlated attribute sets for each column
for col, attr_set in correlated_attributes.items():
    print(f"Correlated attributes for {col}: {attr_set}")

# Step 8: (Optional) Save the correlated attribute sets for future use
#correlated_df = pd.DataFrame({col: list(attrs) for col, attrs in correlated_attributes.items()})
#correlated_df.to_csv('correlated_attributes.csv', index=False)
```

```python
import pandas as pd
import time
from collections import defaultdict

# Define the attributes and constants
attributes = ['SchoolName', 'DistrictCode']
min_support_list = [0.05]
conf_min_list = [0.2]
data_size = []
time_complexity_fpgrowth3 = []
len_of_rules_list_fp1_3 = []
len_of_rules_list_fp2_3 = []
len_of_rules_list_fp3_3 = []
len_of_rules_list_fp4_3 = []
len_of_rule = 0
sorted_filter_rule_list = []

from fpgrowth_py import fpgrowth

# Define the attributes and constants
attributes = ['DistrictCode', 'SchoolCode', 'SchoolName', 'ClassID', 'ClassCourseID']
min_support_list = [0.2]
conf_min_list = [0.15]

# Store rules with corresponding column names and correlation score
sorted_filter_rule_list = []


from fpgrowth_py import fpgrowth

# Iterate over min_support values
for min_support in min_support_list:
    columns_to_use = ["CourseCode"]
    columns_to_use.extend(attributes)
    dataset_new = dataset_to_use.loc[:, columns_to_use]
    dataset_reduced = dataset_new.head(200000)  # Use 200,000 data size, but adjust as needed
    original_dataset = dataset_reduced
    data_group = dataset_reduced.groupby('CourseCode')

    # Get the column mapping dictionary
    column_mapping = defaultdict(lambda: 'Value')
    for column in dataset_reduced.columns:
        column_mapping.update({value: column for value in dataset_reduced[column].unique()})

    # Iterate over confidence values
    for conf_min in conf_min_list:
        len_of_rule = 0
        antecedent_values = []
        consequent_values = []
        # Iterate over groups in the dataset
        for group in data_group.groups:
            dataset_reduced = data_group.get_group(group)
            itemSetList = dataset_reduced.values.tolist()
            freqItemSet, rules = fpgrowth(itemSetList, minSupRatio=min_support, minConf=conf_min)

            # Iterate over rules and store them in a list of dictionaries
            rules_data = []

            # Define the minimum and maximum item count for a rule
            min_item_count = 4  # Adjust as needed
            max_item_count = 4  # Adjust as needed
            for rule in rules:
                antecedent_set, consequent_set, correlation_score = rule

                # Check for NaN and perfect correlation
                if (
                    any("nan" in item for item in (antecedent_set, consequent_set))
                    or correlation_score == 1.0
                ):
                    continue

                # Combine antecedent and consequent sets into a single list of values
                rule_values = list(antecedent_set.union(consequent_set))
                antecedent_values.append(list(antecedent_set))
                consequent_values.append(list(consequent_set))

                if len(rule_values) < max_item_count:
                  continue


                if min_item_count <= len(rule_values) <= max_item_count:

                  len_of_rule = len_of_rule + len(rule_values)


                # Create a dictionary to store the rule values with correct column names
                rule_data = {}

                # Iterate over rule_values and assign them to their corresponding columns

                for value in rule_values:
                    # Map the value to the corresponding column name
                    column_name = column_mapping[value]
                    rule_data[column_name] = value

                # Add the correlation score to the dictionary
                rule_data["Correlation_Score"] = correlation_score

                # Append the rule data to the list
                rules_data.append(rule_data)

            # Create DataFrame from the list of dictionaries for each group
            group_rules_pd = pd.DataFrame(rules_data)

            # Append the group specific rules to the overall sorted_rules_pd
            sorted_filter_rule_list.append(group_rules_pd)

        if conf_min == 0.15:
            len_of_rules_list_fp1_3.append(len_of_rule)
            len_of_rule = 0
        if conf_min == 0:
            len_of_rules_list_fp2_3.append(len_of_rule)
            len_of_rule = 0
        if conf_min == 0:
            len_of_rules_list_fp3_3.append(len_of_rule)
            len_of_rule = 0
        if conf_min == 0:
            len_of_rules_list_fp4_3.append(len_of_rule)
            len_of_rule = 0


# Concatenate all the DataFrames in sorted_filter_rule_list
sorted_rules_pd = pd.concat(sorted_filter_rule_list, ignore_index=True)

# Save the DataFrame to a CSV file
sorted_rules_pd.to_csv("sorted_fillter_rule_course_list32.csv", index=False)
```

```python
# Preprocess the data
['DistrictCode',
 'SchoolCode',
 'DistrictName',
 'SchoolName',
 'ClassID',
 'CourseCode',
 'ClassCourseID',
 'EnrollTotal']
records = []
for i in range(0,len(dataset_to_use.loc[:,["DistrictName"]])):
  record = []
  for j in range(1,2):
    val = dataset_to_use.values[i, j].split(",")
    for k in val:
      if k not in record:
        record.append(k)
  records.append([i, record])

print(record)
# Create the pandas DataFrame
out = pd.DataFrame(records, columns = ['id', 'items'])
out.head()
```

```python
# Convert the pandas dataframe into spark dataframe
sqlContext = SQLContext(spark)
sf = spark.createDataFrame(out,["id", "items"])
sf.show()
```

```python
# Create fpGrowth object by instantiating the FPGrowth constructor with the necessary parameters
fpGrowth = FPGrowth(itemsCol="items", minSupport=0.01, minConfidence=0.01)

# Fit the dataframe into the fpGrowth object to get the model prepared
model = fpGrowth.fit(sf)
# Display the frequently occuring items
model.freqItemsets.show()
```

```python
# Display the generated association rules
association = model.associationRules
association.show()
```

```python
association_pandas = association.toPandas()

#Mean
confidence_mean = mean(association_pandas['confidence'].tolist())
support_mean = mean(association_pandas['support'].tolist())
lift_mean = mean(association_pandas['lift'].tolist())

#Varainace
confidence_variance = variance(association_pandas['confidence'].tolist())
support_variance = variance(association_pandas['support'].tolist())
lift_variance = variance(association_pandas['lift'].tolist())

#Median
confidence_median = median(association_pandas['confidence'].tolist())
support_median = median(association_pandas['support'].tolist())
lift_median = median(association_pandas['lift'].tolist())

#Standard Deviation
confidence_stdev = stdev(association_pandas['confidence'].tolist())
support_stdev = stdev(association_pandas['support'].tolist())
lift_stdev = stdev(association_pandas['lift'].tolist())
```
